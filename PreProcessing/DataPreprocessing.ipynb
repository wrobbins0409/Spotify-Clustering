```python
# imports
import pandas as pd
import csv
import numpy as np
from google.colab import files
```

# Load datasets and show their features

```python
# Load spotify dataset
spotify = pd.read_csv("spotify.csv") # source: https://www.kaggle.com/datasets/joebeachcapital/top-10000-spotify-songs-1960-now/data

# show size
print(spotify.shape)

# display dataframe
display(spotify)
```

```python
# Load tiktok dataset
tiktok = pd.read_csv("tiktok.csv") # source: https://www.kaggle.com/datasets/sveta151/tiktok-popular-songs-2019?resource=download

# show size
print(tiktok.shape)

# display datafraom
display(tiktok)
```

# Ensure identical features have same names

```python
# lower all spotify column names
spotify.columns = map(str.lower, spotify.columns)

# lower all tiktok column names
tiktok.columns = map(str.lower, tiktok.columns)

print(", ".join(spotify.columns))
print(", ".join(tiktok.columns))

# spotify has two features "track name", "artist name(s)" and tiktok has two features "track_name", "artist_name" that are identical

# rename spotify features to match
spotify.rename(columns={'track name': 'track_name', 'artist name(s)': 'artist_name'}, inplace=True)
```

# Join datasets based on which features they share

```python
# check common columns
a = np.intersect1d(spotify.columns, tiktok.columns)

# print result
print(", ".join(a))
```

```python
# put both datasets in list
dataframes = [spotify, tiktok]

# join datasets by common columns
spotify_tiktok = pd.concat(dataframes,join='inner', ignore_index=True)

# show new dataset shape
print(spotify_tiktok.shape)
```

# Remove any duplicate entries

```python
# strip and lower track and artist names in order to get rid of any discrepancies
spotify_tiktok['track_name'] = spotify_tiktok['track_name'].str.strip().str.lower()
spotify_tiktok['artist_name'] = spotify_tiktok['artist_name'].str.strip().str.lower()

# remove duplicate rows (if track and artist are the same)
spotify_tiktok.drop_duplicates(subset=['track_name', 'artist_name'], inplace=True, keep="first")

# show new dataset shape
print(spotify_tiktok.shape)
```

# Remove any entries with empty values in them

```python
# Check for NaN values in the DataFrame
nan_values = spotify_tiktok.isna().sum()

# print nan values
print(nan_values)

# Remove entries with empty values
spotify_tiktok = spotify_tiktok.dropna()

# show new dataset shape
print(spotify_tiktok.shape)
```

# Add sources to entries so that we can check if they are from spotify or tiktok

```python
# Create a set of lowercased and stripped track names for each platform
spotify_tracks_set = set(spotify['track_name'].str.strip().str.lower())
tiktok_tracks_set = set(tiktok['track_name'].str.strip().str.lower())

# Check if each track is from Spotify or TikTok
spotify_tiktok['spotify'] = spotify_tiktok['track_name'].str.strip().str.lower().isin(spotify_tracks_set)
spotify_tiktok['tiktok'] = spotify_tiktok['track_name'].str.strip().str.lower().isin(tiktok_tracks_set)

# show source columns
sources = ['tiktok', 'spotify']

display(spotify_tiktok[sources])
```

```python
# Only print the Spotify entries
spotify_entries = spotify_tiktok[(spotify_tiktok['spotify'] == True) & (spotify_tiktok['tiktok'] == False)]
print(spotify_entries.shape)

# Only print the TikTok entries
tiktok_entries = spotify_tiktok[(spotify_tiktok['spotify'] == False) & (spotify_tiktok['tiktok'] == True)]
print(tiktok_entries.shape)

# Show entries where the song is popular on both Spotify and TikTok
both_entries = spotify_tiktok[(spotify_tiktok['spotify'] == True) & (spotify_tiktok['tiktok'] == True)]
print(both_entries.shape)
```

# Save new dataset to a .csv file

```python
# export dataset
spotify_tiktok.to_csv('spotify_tiktok.csv', index=False)
files.download('spotify_tiktok.csv')
```
